```python
#import dependencies
import pandas as pd
import datetime
import json
import pprint as pp
import csv
from flatten_json import flatten
```

```python
#connect to location history json downloaded from fb and google
fb_data = "json/location_history.json"
```

```python
with open(fb_data ,encoding="utf8") as f:
    fbdata = json.load(f)
    pp.pprint(fbdata)
```

```python
#convert to string
fbstringdata = json.dumps(fbdata)

```

```python
#output to csv

parsed = json.loads(fbstringdata)

fb_data = parsed['location_history']

fb_data_file  = open('fb_data_file.csv', 'w')

csvwriter = csv.writer(fb_data_file)

count = 0

for data in fb_data:

      if count == 0:

             header = data.keys()

             csvwriter.writerow(header)

             count += 1

      csvwriter.writerow(data.values())

fb_data_file.close()
```

```python
#import csv as df
fb_data_csv = ("fb_data_file.csv")
fb_df = pd.read_csv(fb_data_csv)
fb_df.head()
```

```python
#split coordinate into seperate columns 

# new data frame with split value columns 
fixed_df = fb_df["coordinate"].str.split(",", n = 1, expand = True) 

# making separate first name column from new data frame 
fb_df["latitude"]= fixed_df[0] 
  
# making separate last name column from new data frame 
fb_df["longitude"]= fixed_df[1]

fb_df.head()
```

```python
fb_df['latitude'] = fb_df['latitude'].str.replace("{'latitude':", '')
```

```python
fb_df['longitude'] = fb_df['longitude'].str.replace("'longitude':", '')
```

```python
fb_df['longitude'] = fb_df['longitude'].str.replace("}", '')
```

```python
fb_df.head()
```

```python
fb_df_final = fb_df.rename(columns={'creation_timestamp':'timestamp', 'name':'city'})
```

```python
fb_df_final.head()
```

```python
fb_df_final = fb_df_final.drop(columns={'coordinate'})
```

```python
fb_df_final.head()
```

```python
fb_df_final['timestamp'] = pd.to_datetime(fb_df_final['timestamp'],unit='s')
```

```python
fb_df_final.head()
```

```python
#export pd to csv
fb_df_final.to_csv("fb_clean_data.csv", index=False)
```
